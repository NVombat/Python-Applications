```python
#Add a new column with header "Continent" and list of values which will fill the row
df5["Continent"] = ["North America", "Asia", "Africa", "Antarctica", "South America", "Europe"]
df5
```

```python
#Add a new column with header "Status"
#df5.shape returns the number of rows and columns (r,c)
#df5.shape[0] returns the number of rows 
#Thus we set the data to be "CLOSED" for all rows in that column
df5["Status"] = df5.shape[0]*["Closed"]
df5
```

```python
#Updates value of the column to have value of country column and new string
df5["Continent"] = df5["Country"] + "," + "North America"
df5
```

```python
#Transposes the dataframe -> Thus rows become columns and vice versa
#With this method you can add rows to the dataframe
df5_t = df5.T
df5_t
```

```python
#Add a new row but as a column
df5_t[7] = ["7", "Shakar Vihar", "Delhi", "DL 110010", "India", "Swish", "69", "India Asia", "Open"]
df5_t
```

```python
#Transpose the dataframe again to get the new row in the correct place 
df5 = df5_t.T
df5
```

```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
#Returns address with PINCODE, Address, Lat and Long if location exists else returns None
n = nom.geocode("Shankar Vihar, Delhi, 110010")
print(n)
```

```python
#Print latitude and longitude
print(n.latitude)
print(n.longitude)
```

```python
#Modify address column to store entire address with city, state, country and ZIP code
df5["Addr"] = df5["Addr"] + ", " + df5["City"] + ", " + df5["State #"] + ", " + df5["Country"] 
df5
```

```python
#Apply the geocode function to get the data for those addresses
#.apply() -> applies the function to all the elements in the data frame
df5["Coordinates"] = df5["Addr"].apply(nom.geocode)
df5
```

```python
#List of data added to new column
list(df5["Coordinates"])
```

```python
#Create two columns for lat and long -> Apply the latitude and longitude functions using lambda function
#We cater for none type objects as they throw an error
df5["Latitude"] = df5["Coordinates"].apply(lambda x:x.latitude if x!=None else None)
df5["Longitude"] = df5["Coordinates"].apply(lambda x:x.longitude if x!=None else None)
df5
```
